```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

test_dir = './chest_xrays/chest_xray/test'
train_dir = './chest_xrays/chest_xray/train'
train_normal_dir = os.path.join(train_dir, 'NORMAL')
train_pneumonia_dir = os.path.join(train_dir, 'PNEUMONIA')

# counting number of images

num_train_normal = len(os.listdir(train_normal_dir))
num_train_pneumonia = len(os.listdir(train_pneumonia_dir))

print("-" * 60)
print(f"Number of NORMAL training images: {num_train_normal}")
print(f"Number of PNEUMONIA training images: {num_train_pneumonia}")
print("-" * 60)
```

```python
# load and display example images
# file path for each type
normal_img_path = os.path.join(train_normal_dir, os.listdir(train_normal_dir)[0])
pneumonia_img_path = os.path.join(train_pneumonia_dir, os.listdir(train_pneumonia_dir)[0])

# load the images using cv2
normal_img = cv2.imread(normal_img_path)
pneumonia_img = cv2.imread(pneumonia_img_path)

# figure with two subplots to show images side by side 
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# display the normal image
axes[0].imshow(normal_img, cmap='gray') # gray for medical imgs
axes[0].set_title('Normal X-ray')
axes[0].axis('off') # hide axes

# display the pneumonia img 
axes[1].imshow(pneumonia_img, cmap='gray')
axes[1].set_title('Pneumonia X-Ray')
axes[1].axis('off')

plt.show()
```
